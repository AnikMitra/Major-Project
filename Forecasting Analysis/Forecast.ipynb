# **Forecasting for India**

```python
import pandas as pd
df=pd.read_excel('/content/drive/MyDrive/MajorProject/india_data.xlsx')
```

```python
df.tail()
```

# **Forecasting for new daily cases for India**

```python
df1=df[['date','new_cases']]
```

```python
df1 = df1.rename(columns={df1.columns[0]:'ds',
                        df1.columns[-1]:'y'})
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['ds'] = pd.to_datetime(df1['ds'])
df1['y'] = df1['y'].astype(int)
```

```python
from fbprophet import Prophet
import matplotlib.pyplot as plt
model = Prophet(changepoint_prior_scale=0.2)
model.fit(df1)
x = model.make_future_dataframe(periods=30,freq='D')
forecast = model.predict(x)
plt.figure(figsize=(15,7))
plt.plot(forecast['ds'],forecast['yhat'])
```

```python

model.plot(forecast,figsize=(15,7));
```
